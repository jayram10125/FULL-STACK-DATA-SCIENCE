### Creating Dataframe
- simply 2D data consisting rows and columns is called as Dataframe and each column is called as Series
-  DataFrame ke each row ko bhi series kaha jata hai

```python
# using lists
# agar aapko ek dataframe banana hai to 2d list ka need padega
# simply ek list banana chahate hain jisme ki store karenge IQ,Makrs and package 
student_data=[
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]

pd.DataFrame(student_data,columns=["iq",'marks','package'])
```

```python
student_data=[
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]

pd.DataFrame(student_data)
```
