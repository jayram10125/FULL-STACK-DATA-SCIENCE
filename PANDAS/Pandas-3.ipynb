```python
# on dataframe
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
# suppose dono point ke bich ka euclidean distance nikal kar ek alag column me store karna hai
# hum row ke upar bhi apply chala sakte hain

def euclidean(row):
    pt_A = row['1st point']
    pt_B = row['2nd point']

    return ((pt_A[0]-pt_B[0])**2 + (pt_A[1]-pt_B[1])**2)**0.5
```

```python
points_df.apply(euclidean,axis=1)   # axis = 1 means oassing a row
```

```python
points_df['distance']=points_df.apply(euclidean,axis=1)
points_df
```
