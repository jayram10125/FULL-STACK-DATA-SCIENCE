### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot


example-scatter plot , line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python

tips=sns.load_dataset('tips')
tips
```

```python
# scatter plot   -> axis level function
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
# relplot-> figure level -> generally in square shape(default but hum shape change kar kar sakte hain)
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
# accordiny to seaborn library it is suggested ki hume figure level jyada use karna chahaiye qki isme sirf kind change karke hum multiple 
# types ke graph plot kar sakte hain  bahut sare function yaad nahi rakhna parta hai
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='time',style='sex') # hue style ko ulta pulta karo koi matlab nahi
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
# plotting univariate histogram
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
# bins parameter
sns.displot(data=tips,x='total_bill',kind='hist',bins=20)
```

```python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables
#(generally histogram se  single type ke data anlysis kya jata hai but seaborn me ye option hai ki hum categorical column pe bhi ye kaam kar sakte hain)

# countplot(conventionally ye count plot hai but hum histogram kah sakte hain# )
```

```python
sns.displot(data=tips,x='day',kind='hist')
```

```python
#hue parameter
sns.displot(data=tips,x='tip',kind='hist',hue='sex')
```

### kdeplot
- kernal density estimation
- Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate


```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue -> fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
# increasing figure level me size
plt.figure(figsize=(15,7))
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
# size me kuch change nahi hua so isse aaise nahi kar sakte hain
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=10)
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=2)  # aspect ka concept documentation se padho
```

### Rugplot

- Plot marginal distributions by drawing ticks along the x and y axes.

- This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

- (ye plot individual nahi kisis another graph ke sath use hota hai)


```python
sns.rugplot(data=tips,x='total_bill',)
```

```python
sns.kdeplot(data=tips,x='total_bill',)
sns.rugplot(data=tips,x='total_bill',)
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color

sns.histplot(data=tips, x='total_bill', y='tip')
```

```python
# isme vertical and horizontal histogram bana and jaha pe intesect hone ke baad colour dark hai wo jyada quantity ko show kar raha hai like ki 15$ total
# bill and 3$ ka tip jyada logon ke dwara dya gya hai

# hum same kaam displot se bhi kar sakte hain
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

```python
# # Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip',)
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix


# har country ka har saal life expectenties kitna hai
temp_df=gap.pivot(index='country',columns='year',values='lifeExp')
temp_df
```

```python
# heatmap-> axes level function and isse draw karne ka koi bhi figure level function nahi hota hai
sns.heatmap(temp_df)
```

```python
# you can customize the size jaise ki matplotlib me karte the simalar type se
plt.figure(figsize=(15,10))
sns.heatmap(temp_df)
```

```python
# annot  -> a notation
# suppose sirf europe ka heatmap karna hai
temp_df=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
temp_df
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(temp_df)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(temp_df,annot=True)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(temp_df,annot=True,linewidth=0.5)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(temp_df,annot=True,linewidth=0.5,cmap='summer')
```

```python
# Clustermap
#(similar to heatmap but ye internally data ke upar clustring karta hai means ki similar column ko ek jagah laa deta hai 
#  dissimilar data ko alag kar deta HAI)
# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

iris=px.data.iris()
iris

# hume pata karna hai ki inn 4 column mein (sepal_length	sepal_width	petal_length	petal_width	) kya similiraty and kya dissimilarity hai
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
